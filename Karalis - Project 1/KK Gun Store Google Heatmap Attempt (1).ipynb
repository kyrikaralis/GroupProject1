```python
import gmaps
import pandas as pd
import requests

# Google developer API key
from config import google_api

# Configure gmaps
gmaps.configure(api_key=google_api)
```

```python
 # Create aiport dataframe
address_df = pd.read_csv("/Users/kyriakikaralis/Downloads/gunshop_info.csv")
address_df.dropna()
address_df.head()
```

```python
# Store address in locations
address_df["Full_Address"] = address_df["Premise Street"] + address_df["Premise City"] + ", " + address_df["Premise State"]

address_df["Full_Address"]
```

```python
#define locations as the full address
locations = address_df["Full_Address"]
locations.head()
```

```python
#create a For loop so that for every unique address, it runs through the google map (like the open weather API)
url = "https://maps.googleapis.com/maps/api/geocode/json?address="
list = []

for location in locations: 
    try:
        query = f'{url}{location}&key={google_api}'
        response = requests.get(query).json()
        lat= response["results"][0]["geometry"]["location"]['lat']
        lng = response["results"][0]["geometry"]["location"]['lng']
        list.append ({"lat":lat,
                      "lng":lng
        })
        #print (response["results"][0]["geometry"]["location"])
    
    except:
        pass
    
```

```python
 # Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, dissipating=False, max_intensity=10, point_radius=1)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
