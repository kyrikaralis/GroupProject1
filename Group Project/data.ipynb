```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
data = 'full_data.csv'
cdc_df = pd.read_csv(data)

cdc_df
```

```python

gender_groups = cdc_df.groupby('intent')

gender_intent = gender_groups['Unnamed: 0'].count()

gender_chart = gender_intent.plot(kind="pie", figsize = (20,10), title="Gun Deaths Suicide vs Homicide")



plt.ylabel("Total Suicide Count")
plt.savefig("Gun_Deaths_Suicide_vs_Homicide")
plt.show()
plt.tight_layout()
```

```python
import requests
from pprint import pprint
from Config import api_key

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
```

```python
query = "suicide"
```

```python
begin_date = "19990101"
end_date = "20000130"

query_url = f"{url}api-key={api_key}&q={query}&begin_date={begin_date}&end_date={end_date}"
```

```python
articles = requests.get(query_url).json()
articles_list = [article for article in articles["response"]["docs"]]
```

```python
import time

articles_list = []

# loop through pages 0-2
for page in range(0, 20):
    query_url = f"{url}api-key={api_key}&q={query}&begin_date={begin_date}&end_date={end_date}"
    # create query with page number
    query_url = f"{query_url}&page={str(page)}"
    articles = requests.get(query_url).json()
    
    # Add a one second interval between queries to stay within API query limits
    time.sleep(1)
    # loop through the response and append each article to the list
    for article in articles["response"]["docs"]:
        articles_list.append(article)
```

```python
for article in articles_list:
    print({article["snippet"]})
```
